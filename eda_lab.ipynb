11) Write DataFrame to CSV, Excel, and JSON

```python
# Use df_json or any dataframe you want
df_output = pd.DataFrame({
    "Product": ["Book", "Pen", "Notebook"],
    "Price": [50, 10, 25]
})

# Write to CSV
df_output.to_csv("output.csv", index=False)

# Write to Excel
df_output.to_excel("output.xlsx", index=False)

# Write to JSON
df_output.to_json("output.json", orient="records")

print("Files written successfully: output.csv, output.xlsx, output.json")
```

12) Create Random Float Dataset

```python
from numpy.random import randn

df = pd.DataFrame(
    randn(1000),
    index=pd.date_range("2019-06-07", periods=1000),
    columns=["value"]
)

print(df.head())
```

13) Cumulative Sum

```python
df["value"] = df["value"].cumsum()

print("Data after cumulative sum:")
print(df.head())
```

14) Descriptive Statistics

```python
print("Descriptive Statistics:")
print(df.describe())
```

15) Simple Line Plot

```python
df.plot(figsize=(10, 5), title="Cumulative Sum Line Plot")
plt.xlabel("Date")
plt.ylabel("Value")
plt.show()
```

17) Load Iris Dataset

```python
iris = sns.load_dataset("iris")

print(iris.head())
print("\nShape:", iris.shape)
```

18) Describe Iris Dataset

```python
print("Iris Descriptive Statistics:")
print(iris.describe())
```

19) Line Plot for Iris Dataset

```python
iris.plot(figsize=(12, 6), title="Iris Dataset Line Plot")
plt.show()
```

20) Bar Chart

```python
iris.drop(["species"], axis=1).iloc[0].plot(kind="bar", title="Bar Chart of First Iris Record")
plt.show()
```

21) Histogram

```python
iris.plot(kind="hist", bins=20, alpha=0.7, figsize=(10, 6), title="Histogram of Iris Features")
plt.show()
```

22) Distribution Plot

```python
iris["sepal_length"].plot(kind="hist", bins=20, title="Distribution of Sepal Length")
plt.xlabel("Sepal Length")
plt.show()
```

23) Scatter Plot

```python
iris.plot.scatter(
    x="sepal_length",
    y="petal_length",
    figsize=(8, 6),
    title="Scatter Plot: Sepal Length vs Petal Length"
)
plt.show()
```

24) Scatter Plot with Color Dimension

```python
iris.plot.scatter(
    x="sepal_length",
    y="petal_length",
    c="sepal_width",
    s=80,
    figsize=(8, 6),
    title="Scatter Plot with Color"
)
plt.show()
```

25) Hexbin Plot

```python
iris.plot.hexbin(
    x="sepal_length",
    y="petal_length",
    gridsize=10,
    figsize=(8, 6),
    title="Hexbin Plot"
)
plt.show()
```

26) Scatter Matrix

```python
from pandas.plotting import scatter_matrix

scatter_matrix(
    iris.drop("species", axis=1),
    figsize=(10, 10),
    diagonal="hist"
)
plt.show()
```

27) KDE Plot

```python
df["value"].plot.kde(figsize=(8, 5), title="KDE Plot")
plt.show()
```

28) Subplots

```python
iris.drop("species", axis=1).plot(
    subplots=True,
    layout=(2, 2),
    figsize=(12, 8),
    title="Subplots of Iris Features"
)
plt.tight_layout()
plt.show()
```

29) Pie Plot

```python
iris.drop("species", axis=1).iloc[0].plot.pie(figsize=(7, 7), autopct="%.2f")
plt.title("Pie Plot of First Iris Row")
plt.ylabel("")
plt.show()
```

30) Andrews Curves

```python
from pandas.plotting import andrews_curves
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
andrews_curves(iris, "species")
plt.title("Andrews Curves for Iris Dataset")
plt.show()
```
